```
import pandas as pd 
import datetime
import numpy as np
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
#Chirag
fundamentals = pd.read_csv("drive/MyDrive/Colab Notebooks/fundamentals.csv")
fundamentals
```

```
fundamentals["Period Ending"] = pd.to_datetime(fundamentals["Period Ending"])
days = datetime.timedelta(1)
year = datetime.timedelta(364)
fundamentals["PeriodAdjusted"] = fundamentals["Period Ending"] - days
fundamentals["PeriodLagged"] = fundamentals["Period Ending"] + year
fundamentals
```

```

prices = pd.read_csv("drive/MyDrive/Colab Notebooks/prices-split-adjusted.csv")
prices
```

```
prices["date"] = pd.to_datetime(prices["date"])
prices
```

```
joined = pd.merge(fundamentals, prices, left_on=["Ticker Symbol", "PeriodAdjusted"], right_on=["symbol","date"], how="inner")
#vars = ["PeriodAdjusted", "date","symbol", "open","low","high","volume"]
joined
```

```
joined_lag = pd.merge(joined, prices, left_on=["Ticker Symbol", "PeriodLagged"], right_on=["symbol","date"], how="inner")
joined_lag
```

```
joined_lag["priceChange"] = joined_lag["close_y"]/joined_lag["close_x"]
joined_lag["priceChange"].describe()
```

```
spPerformance = {"Year":[2012, 2013, 2014,2015,2016], "Market Performance": [1.2960, 1.1139, 0.9927, 1.0954, 1.1942]}
marketPerf = pd.DataFrame(data=spPerformance)
marketPerf
```

```
total_data = pd.merge(joined_lag, marketPerf, left_on=["For Year"], right_on=["Year"])
total_data["perc_performance"] = total_data["priceChange"]-total_data["Market Performance"]
```

```
total_data["perc_performance"].describe()
```

# Response Variable Encoding: 
**Good (Outperformance):** [.06, inf)

**Neutral (Neither outperformed nor underperformed significantly):** [-0.06, 0.06)

**Bad (Underperformance):** [-inf, -0.06)


```
total_data["performance"] = pd.cut(total_data["perc_performance"], bins=[-1000,-0.059,0.06,1000], labels=("Underperformed", "Neutral", "Outperformed"))
total_data["performance"].value_counts()
```

```
total_data["Receivables Turnover"] = total_data["Total Revenue"]/total_data["Accounts Receivable"]
total_data["Inventory Turnover"] = total_data["Cost of Revenue"]/total_data["Inventory"]
total_data["Debt Ratio"] = total_data["Total Liabilities"]/total_data["Total Assets"]
total_data["Return on Assets"] = total_data["Net Income"]/total_data["Total Assets"]
```

```
vars = ["After Tax ROE", "Cash Ratio", "Current Ratio", "Gross Margin", "Operating Margin", "Pre-Tax Margin", "Pre-Tax ROE", "Profit Margin", "Quick Ratio", "Receivables Turnover", "Inventory Turnover", "Debt Ratio", "Return on Assets", "performance"]
```

```
final_data = total_data[vars]
final_data_v1 = final_data[~final_data.isin([np.nan, np.inf, -np.inf]).any(1)] #dropped nas and infinity values
```
